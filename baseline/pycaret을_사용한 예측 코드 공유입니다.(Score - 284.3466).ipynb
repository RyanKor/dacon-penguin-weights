# 1.라이브러리 불러오기

```python
import os
os.chdir('/content/drive/MyDrive/dacon/penguin')

import pandas as pd
import numpy as np
from pycaret.regression import *
```

# 2.데이터 불러오기
- 목표 : Body Mass (g) 예측

```python
train = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')

train = train.drop('id', 1)
train.head()
```

# 3.Pycaret

## 3.1 초기화

```python
setup(train, target = 'Body Mass (g)', session_id = 1234) 
```

## 3.2 모델 비교 
기본 하이퍼 파라미터를 사용해 모델 라이브러리의 모든 모델을 훈련하고 교차 검증을 사용해 성능 메트릭을 평가하며 이는 훈련된 모델 객체를 반환한다.

사용된 평가 지표는 다음과 같다. 
* Classifiacation(분류)  
 -  Accuracy, AUC, Recall, Precision, F1, Kappa, MCC
* Regression(회귀) 
 - MAE, MSE, RMSE, R2, RMSLE, MAPE

```python
compare_models(sort = 'RMSE')
```

## 3.3 상위 3개 모델 생성

```python
llar = create_model('llar', cross_validation = False)
lasso = create_model('lasso', cross_validation = False)
lr = create_model('lr', cross_validation = False)
```

## 3.4 하이퍼파라미터 조정

```python
tune_llar = tune_model(llar, optimize = 'RMSE', n_iter = 100)
tune_lasso = tune_model(lasso, optimize = 'RMSE', n_iter = 100)
tune_lr = tune_model(lr, optimize = 'RMSE', n_iter = 100)
```

```python
blended_specifier = blend_models(estimator_list = [tune_llar, tune_lasso, tune_lr], fold = 10, optimize = 'RMSE')
```

## 3.5 모델 배포

```python
final_model = finalize_model(blended_specifier)
```

# 4.제출

```python
submission = pd.read_csv('sample_submission.csv', index_col = 0)
submission.head()
```

```python
test = test.drop('id', 1)
test.head()
```

```python
pred = predict_model(final_model, data = test)
pred.head()
```

```python
submission['Body Mass (g)'] = pred['Label']
submission.head()
```

```python
submission.to_csv('pycaret.csv')
```
